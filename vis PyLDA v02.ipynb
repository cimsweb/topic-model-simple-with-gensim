```python
# -*- coding: utf-8 -*-
import gensim
from gensim import corpora, similarities, models, matutils
import time
import re
import numpy as np
import pandas as pd
import logging
import html

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
pyLDAvis.enable_notebook()

import string
```

```python
%%time
lines = []
filename = 'data/Biblia_v_formate_rtf.txt'
with open(filename) as file:
    lines = [line.rstrip() for line in file]
    
print (len(lines))

list_out_big = []
stoplist = set('в и же не как что так почему они его вас для вам кто мне нас них или но'.split())
for row in lines:
    out_text = []
    for word in row.split():
        w = word.strip(string.punctuation).lower()
        if len(w)>2 and w not in stoplist:
            out_text.append(word.strip(string.punctuation).lower())
    list_out_big.append(out_text)
    
print (len(list_out_big))

dictionary = corpora.Dictionary(list_out_big)

corpus = [dictionary.doc2bow(text) for text in list_out_big]

lda = gensim.models.ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=7, alpha='auto', eval_every=5)
lda_viz = gensimvis.prepare(lda, corpus, dictionary)
print (len(dictionary))
```
